```python
def handle_outlier(df,cols,method="IQR",threshold=1.5,strategy="replace"):
  for col in cols:
    if method=="IQR":
      Q1=df[col].quantile(0.25)
      Q3=df[col].quantile(0.75)
      IQR = Q3-Q1
      lower_bound = Q1-threshold*IQR
      upper_bound = Q3+threshold*IQR
    elif method=="Z-score":
      mean=df[col].mean()
      std=df[col].std()
      lower_bound=mean-threshold*std
      upper_bound=mean+threshold*std
    else:
      print("Invalid method")
    if strategy=="replace":
      df[col]=np.clip(df[col],lower_bound,upper_bound)
    elif strategy=="remove":
      df=df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]
    else:
      print("Invalid strategy. shoose 'replace' or 'remove'")
  return df
column_to_check=[["meals_served","food_waste_kg"]]



column_to_check=df[["temperature_C"]]
df=handle_outlier(df,column_to_check,method="IQR",threshold=.75,strategy="replace")
```
